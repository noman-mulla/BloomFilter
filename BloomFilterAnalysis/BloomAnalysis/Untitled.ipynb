```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loadData():
    train_file_path = "output.csv"
    train_data = pd.read_csv(train_file_path)
    return train_data
```

```python
train_data = loadData()
train_data.head()
```

```python
plt.figure(figsize=(15,10))
sns.set(font_scale=2)
sns.factorplot(x="FalsePosRate", y="HashFunctions", data = train_data, kind="bar", hue="BloomFilterSize",
              size=15, aspect=2, palette='coolwarm')
```

```python
train_data = train_data.sort_values(by=['HashFunctions'])
train_data.head()
```

```python
train_data['FalsePosRate'] = train_data['FalsePosRate'].apply(lambda x: x*100)
train_data.columns
```

```python
train_data['Hash Combination Number'] = np.arange(len(train_data))
```

```python
plt.figure(figsize=(200,200))
sns.set(font_scale=3)
palette ={18:"red",45:"green",90:"blue"}

sns_plot = sns.factorplot(x="FalsePosRate", y="HashFunctions", data = train_data, kind="bar", hue="BloomFilterSize",
              size=60, aspect=1, palette="deep")
sns_plot.savefig("output.png")
```

```python
plt.figure(figsize=(25,15))
sns_scatter = sns.scatterplot(x="Hash Combination Number", y="FalsePosRate", sizes=(100, 200),
                     data=train_data,x_bins=None,s=200)
plt.savefig("output_scatter_2.png")
sns.set(font_scale=3)
```
